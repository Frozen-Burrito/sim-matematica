## Descripción del dataset
Se presenta un conjunto de datos proveniente de investigaciones y estudios experimentales en nanotecnología y ciencia de materiales. Estos se utilizan para analizar las propiedades de los átomos, sus estructuras afectany sus afectaciones a diversas propiedades físicas y químicas.

## Descripción de las variables

- **atom_count**: Cantidad de átomos en la nanopartícula. Tipo de dato: Entero.
- **structure_type**: Tipo de estructura del átomo o material. Tipo de dato: String.
- **atomic_number**: Número atómico del elemento químico (78 Platino, 79 Oro). Tipo de dato: Entero.
- **electronegativity**: Medida de la tendencia de un átomo a atraer electrones en un enlace químico. Tipo de dato: Flotante.
- **atomic_radius**: Medida del radio atómico, expresada en Angstroms. Tipo de dato: Flotante.
- **melting_point**: Punto de fusión del material, expresado en Kelvin. Tipo de dato: Flotante.
- __dos_pc*n*__: Componentes principales de Densidad de Estado (DOS) (n = 1, 2,...,50). Tipo de dato: Flotante.
- **target_property**: Propiedad objetivo que se busca predecir. Tipo de dato: Flotante.

## Visualización de datos

A continuación se presentan visualizaciones estadísticas básicas de las columnas principales del conjunto de datos. Las variables numéricas `atom_count`, `electronegativity`, `atomic_radius` y `melting_point` se describen en histogramas, mientras que la variable categórica `structure_type` se visualiza usando un gráfico de columnas.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Usar value_counts para obtener la frecuencia absoluta de cada tipo de estructura en los datos. 
conteo_estructuras = df["structure_type"].value_counts()

# Crear un gráfico de columnas con los tipos de estructura.
plt.figure(figsize=(8, 6))
plt.bar([str(estructura) for estructura in conteo_estructuras.index], conteo_estructuras.values, color=["orange", "blue", "green"])
plt.xlabel("Tipo de estructura cristalina")
plt.ylabel("Conteo")
plt.title("Distribución de estructuras")

plt.show()
```

```python
from math import ceil

columnas_visualizadas = ["atom_count", "atomic_radius", "electronegativity", "melting_point", "target_property"]

NUM_COLUMNS_IN_FIGURE = 2
NUM_ROWS_IN_FIGURE = ceil(len(columnas_visualizadas) / NUM_COLUMNS_IN_FIGURE)

plt.figure(figsize=(16, NUM_ROWS_IN_FIGURE * 6))

for i, columna in enumerate(columnas_visualizadas):
    plt.subplot(NUM_ROWS_IN_FIGURE, NUM_COLUMNS_IN_FIGURE, i + 1)
    sns.histplot(data=df, x=columna, kde=True, bins=20)
    plt.title(f'Distribution of {columna}')
    sns.despine()

plt.tight_layout()
plt.show()
```
